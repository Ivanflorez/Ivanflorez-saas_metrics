```python
def process_excel_data(filename, columna):
    # Leer el archivo Excel y cargar en DataFrame 'df'
    df = pd.read_excel(filename)

    # Crear DataFrame con la columna especificada
    df_transform = pd.DataFrame(df[columna])

    # Eliminar duplicados en el DataFrame y reiniciar el índice
    df_transform.drop_duplicates(inplace=True)
    df_transform.reset_index(drop=True, inplace=True)

    # Resetear el índice y añadir 1 al índice
    df_transform.reset_index(drop=False, inplace=True)
    df_transform['index'] += 1

    # Cambiar el nombre de la columna 'index' a 'id_nombre_columna'
    df_transform.rename(columns={'index': f'id_{columna}'}, inplace=True)

    return df_transform
```

```python
df_customer = process_excel_data('dataset.xlsx', 'Customer Name')
df_customer
```

```python
df_product = process_excel_data('dataset.xlsx', 'Product Family')
df_product
```

```python
df_loctaion = process_excel_data('dataset.xlsx', 'Location')
df_loctaion
```

```python
df_account = process_excel_data('dataset.xlsx', 'Account Owner')
df_account
```
